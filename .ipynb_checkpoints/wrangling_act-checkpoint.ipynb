## Gathering Data

```python
#Read downloaded twitter archive file by using read csv function.
archive_data = pd.read_csv('twitter-archive-enhanced.csv')

#Download image_predictions file programmatically by using requests library:
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
file_name = "image-predictions.tsv"
r = requests.get(url)
with open(file_name, 'wb') as f:
        f.write(r.content)
image_predictions = pd.read_csv('image-predictions.tsv', delimiter = '\t')
```

```python
#Twitter API Access
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
tweet_ids = archive_data.tweet_id.values

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
```

```python
#Parse json file to create a for loop in the next step
with open('tweet_json.txt') as f:
    tweet_dict = json.loads("[" + 
        f.read().replace("}\n{", "},\n{") + 
    "]")

    print(tweet_dict)

#Create a for loop to retrieve tweet id, like count and retweet count for each tweet.

tweet_list = []
tweet_dict_range = len(tweet_dict)

for i in range(tweet_dict_range):
    tweet = tweet_dict[i]
    ids = tweet['id']
    like_count = tweet['favorite_count']
    retweet = tweet['retweet_count']
    tweet_list.append({'id': ids, 'likes': like_count, 'retweets': retweet})    

twitter_json_df = pd.DataFrame(tweet_list, columns = ['id', 'likes', 'retweets'])
```

##### `archive_data`:  9. Dog types in the table are separate columns. It should be melted into one column.


#### Define
melt doggo, floofer, pupper, puppo types into one column and name it as 'dog stage'

#### Code

```python
# handle nones in the columns
archive_data_clean.doggo.replace('None', '', inplace=True)
archive_data_clean.floofer.replace('None', '', inplace=True)
archive_data_clean.pupper.replace('None', '', inplace=True)
archive_data_clean.puppo.replace('None', '', inplace=True)

# merge into column
archive_data_clean['dog_stage'] = archive_data_clean.doggo + archive_data_clean.floofer + archive_data_clean.pupper + archive_data_clean.puppo

# handle multiple stages
archive_data_clean.loc[archive_data_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper'
archive_data_clean.loc[archive_data_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo'
archive_data_clean.loc[archive_data_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'

# handle missing values
archive_data_clean.loc[archive_data_clean.dog_stage == '', 'dog_stage'] = np.nan

#drop empty doggo, floofer, pupper, puppo columns as we already have the new dog_stage column
archive_data_clean.drop(columns={'doggo','floofer','pupper','puppo'}, inplace = True)
```
